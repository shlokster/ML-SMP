```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D

df = pd.read_csv("Simple_Dataset - Sheet1.csv")
df.head()
```

```python
plt.scatter(df["x"], df["y"])
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4)
kmeans.fit(df)
```

```python
labels = kmeans.predict(df)
centroids = kmeans.cluster_centers_

colmap = {1:'r', 2:'b', 3:'g', 4:'k'}
colors = list(map(lambda x : colmap[x+1], labels))
colors1 = list(colors)
plt.scatter(df['x'], df['y'], color=colors1, alpha=0.5, edgecolor='k')

for idx, centroid in enumerate(centroids):
  plt.scatter(*centroid, color=colmap[idx+1])

plt.show()
```

```python
df = pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
df = df.drop("Gender", axis=1)
df = df.drop("CustomerID", axis=1)
df.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(df.iloc[:, 0], df.iloc[:, 1], df.iloc[:, 2])
ax.set_xlabel(df.columns[0])
ax.set_ylabel(df.columns[1])
ax.set_zlabel(df.columns[2])

plt.show()
```

Now we start Clustering


```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(df)
```

```python
labels = kmeans.predict(df)
centroids = kmeans.cluster_centers_

colmap = {1:'r', 2:'g', 3:'b', 4:'k', 5:'yellow'}

fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111, projection="3d")

colors = list(map(lambda x: colmap[x+1], labels))
colors1=list(colors)
ax.scatter(df.iloc[:, 0], df.iloc[:, 1], df.iloc[:, 2], c=colors1, alpha=0.5, edgecolor='k')

for idx, centroid in enumerate(centroids):
  ax.scatter(*centroid, c=colmap[idx + 1])
plt.show()
```
